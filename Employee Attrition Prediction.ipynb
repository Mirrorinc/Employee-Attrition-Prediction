**Splitting data into test and train**

**Scaling dataset**

```python
print(X_train.shape) 
```

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

# Defining column transformer to handle categorical and numerical variables separately
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), [2, 4, 6, 7, 9, 13, 14, 15, 16, 17]),  # Replacing 'numerical_columns' with the list of numerical column indices
        ('cat', OneHotEncoder(), [0, 1, 3, 5, 8, 10, 11, 12])  # Replacing 'categorical_columns' with the list of categorical column indices
    ])

# Defining the preprocessing pipeline
pipeline = Pipeline(steps=[('preprocessor', preprocessor)])

# Fitting and transform the training data
X_train_processed = pipeline.fit_transform(X_train)

# Transforming the test data
X_test_processed = pipeline.transform(X_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
DTClassifier = DecisionTreeClassifier(criterion = "entropy", random_state = 0)
DTClassifier.fit(X_train_processed, y_train)
```

```python
y_pred = DTClassifier.predict(X_test_processed)
```

```python
y_pred
```

```python
from sklearn import metrics
print("The accuracy of decision tree is : ", metrics.accuracy_score(y_pred, y_test))
```

**Not well satisfatied with the accuracy, let's try naive bayes**

```python
from sklearn.naive_bayes import GaussianNB
NBClassifier = GaussianNB()
NBClassifier.fit(X_train_processed, y_train)
```

```python
y_pred = NBClassifier.predict(X_test_processed)
```

```python
y_pred
```

```python
print("The accuracy of Naive Bayes is: ", metrics.accuracy_score(y_pred, y_test))
```
